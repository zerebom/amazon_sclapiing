```python
import urllib.request as urllib
from bs4 import BeautifulSoup
import re
from contextlib import closing
import sqlite3
import time
```

```python


dbname='1105_review_of_bakawall7.db'
j=0
with closing(sqlite3.connect(dbname)) as conn:
    c = conn.cursor()
    create_table = '''create table reviews (id int, name, date, rate float, url,
                       len int, review text)'''
    c.execute(create_table)

    
    for i in range(1,64):
        print(i)
        url=f'https://www.amazon.co.jp/%E3%83%90%E3%82%AB%E3%81%AE%E5%A3%81-%E6%96%B0%E6%BD%AE%E6%96%B0%E6%9B%B8-%E9%A4%8A%E8%80%81-%E5%AD%9F%E5%8F%B8/product-reviews/4106100037/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber={i}'
        html = urllib.urlopen(url)
        soup = BeautifulSoup(html,'lxml')
        review_area=soup.select('#cm_cr-review_list')[0]
        each_reviews=review_area.select('.a-section.review')
        time.sleep(3)

        for review in each_reviews:
            j+=1
            
            if len(review.select('.a-profile-name'))==0:
                continue
            rate=review.select('.a-icon-alt')[0].text
            rate=float(re.sub('5つ星のうち','',rate))
            
            date=review.select('.review-date')[0].text
            text=review.select('.review-text')[0].text
            name=review.select('.a-profile')[0].text
            title=review.select('.review-title')[0].text
            reviewer_url=review.select('.a-profile')[0].get('href')
            
            sql = 'insert into reviews (id, name, date, rate, url, len, review) values (?,?,?,?,?,?,?)'
            user = (j,date,name,rate,reviewer_url,len(text),text)
            print(id,name)
            c.execute(sql, user)
    conn.commit()

    
#     print(date)
#     print(name)
#     print(title)
#     print(reviewer_url)
        
    
    

# review_rank=review_area.select('.a-icon-alt')
# review_rank=[x.get_text() for x in review_rank]
# review_rank=[float(re.sub('5つ星のうち','',x)) for x in review_rank]
          
```

```python
reviewer_name=[]
reviewer_url=[]
review_title=[]
for i in range(1,64):
    print(i)
    url=f'https://www.amazon.co.jp/%E3%83%90%E3%82%AB%E3%81%AE%E5%A3%81-%E6%96%B0%E6%BD%AE%E6%96%B0%E6%9B%B8-%E9%A4%8A%E8%80%81-%E5%AD%9F%E5%8F%B8/product-reviews/4106100037/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber={i}'
    driver.get(url)
    review_area = driver.find_element_by_id('cm_cr-review_list')
    reviewer_name.extend([x.text for x in review_area.find_elements_by_css_selector('.a-profile')])
    reviewer_url.extend([x.get_attribute('href') for x in review_area.find_elements_by_css_selector('.a-profile')])
    review_title.extend([x.text for x in review_area.find_elements_by_css_selector('.review-title')])
    print(len(reviewer_name),len(reviewer_url),len(review_title))


```
